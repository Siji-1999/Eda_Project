### Identify Total Sales Amount of Each Category

```python
###Total sales amount of each category
total_sales_by_Category=df.groupby('Category')['Sales'].sum().reset_index()
print(total_sales_by_Category)
sns.barplot(x='Category', y='Sales', data=total_sales_by_Category,palette='Set2')
plt.title('Total Sales by Category')
plt.show()
```

Based on the total sales data for each category;
Technology: This category has the highest total sales, amounting to approximately 4.74 million. This indicates a significant demand and robust performance in the technology sector.
Furniture: The total sales for Furniture are approximately 4.11 million, making it the second-highest category. This shows a strong and steady demand for furniture products.
Office Supplies: With total sales of approximately 3.79 million, Office Supplies is the lowest among the three categories. Despite being the lowest, it still represents a considerable portion of the overall sales.

### Identify which city has the highest sales

```python
sales_by_city = df.groupby('City')['Sales'].sum().reset_index()
top_city_idx = sales_by_city['Sales'].idxmax()
top_city = sales_by_city.iloc[top_city_idx]['City']
top_city_sales = sales_by_city.iloc[top_city_idx]['Sales']
print(f"The city with the highest sales is {top_city} with a total of {top_city_sales} in sales")
top_cities = sales_by_city.head(10)

sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
colors = sns.color_palette("viridis", len(top_cities))
plt.pie(top_cities['Sales'], labels=top_cities['City'], autopct='%1.1f%%', colors=colors)
plt.title('Top 10 Cities by Sales')
plt.axis('equal')  
plt.show()


```

It clearly states that Aachen city has the highest  sales compared to other cities

### Most Common Sub-category

```python
most_common_sub_category=df.groupby('Sub-category')['Sales'].sum().reset_index()
print(f"The most common sub-category is{most_common_sub_category}")

sns.set_style("whitegrid")
plt.figure(figsize=(18, 6))
barplot = sns.barplot(x='Sub-category', y='Sales', data=most_common_sub_category,palette='viridis')

plt.title('Most Common Sub-Categories')
plt.xlabel('Sub-category')
plt.ylabel('Sales')
plt.show()
```

##  Identify the Sub-category with the highest profit

```python
profit_by_sub_category = df.groupby('Sub-category')['Profit'].sum().reset_index()
highest_profit_sub_category = profit_by_sub_category.loc[profit_by_sub_category['Profit'].idxmax()]
highest_profit_amount = highest_profit_sub_category['Profit']
highest_profit_name = highest_profit_sub_category['Sub-category']
print(f"The sub-category with the highest profit is {highest_profit_name} with a total of {highest_profit_amount}.")

plt.figure(figsize=(12, 8))
sns.barplot(x='Profit', y='Sub-category', data=profit_by_sub_category, palette='Set2')
plt.title('Total Profit by Sub-Category')
plt.xlabel('Total Profit')
plt.ylabel('Sub-Category')
plt.show()
```

The sub-category with the highest profit is Copiers, generating a total profit of $258567.548. The bar chart visually highlights that Copiers are significantly more profitable compared to other sub-categories, suggesting a potential area for focused investment and marketing efforts.

## State with the highest and lowest Average Profit Per Order

```python
#States with highest and lowest averge profit per order
average_profit_by_state=df.groupby('State')['Profit'].mean().reset_index()
highest_avg_profit_state=average_profit_by_state.loc[average_profit_by_state['Profit'].idxmax()]
lowest_avg_profit_state=average_profit_by_state.loc[average_profit_by_state['Profit'].idxmin()]
print(f"State with highest average profit per order: {highest_avg_profit_state}")
print(f"State with lowest average profit per order: {lowest_avg_profit_state}")

plt.figure(figsize=(10, 6))
sns.barplot(x='State',y='Profit',data=average_profit_by_state.sort_values(by='Profit',ascending=False).head(10),palette='Set2')
plt.title('Top 10 States by Average Profit per Order')
plt.show()
```

## Most common order priority

```python
order_priority_counts = df['Order_priority'].value_counts().reset_index()
order_priority_counts.columns = ['Order_priority', 'Count']
most_common_order_priority = df['Order_priority'].mode()[0]
print(f"The most common order priority is {most_common_order_priority}.")

plt.figure(figsize=(10, 6))
colors = sns.color_palette('Set2', len(order_priority_counts))
plt.pie(order_priority_counts['Count'], labels=order_priority_counts['Order_priority'], autopct='%1.1f%%', colors=colors)
plt.title('Distribution of Order Priorities')
plt.show()
```

The most common order priority is 'Medium', indicating that the majority of orders are assigned this priority level. The pie chart shows that 'Medium' priority orders make up a significant portion of all orders, highlighting its importance in the overall order fulfillment process.








## Identify the order priority distribution vary by customer segment

```python
order_priority_distribution = df.groupby(['Segment', 'Order_priority']).size().reset_index(name='Count')
print(order_priority_distribution)

plt.figure(figsize=(12, 8))
sns.barplot(x='Order_priority', y='Count', hue='Segment', data=order_priority_distribution, palette='Set2')
plt.title('Order Priority Distribution by Customer Segment')
plt.xlabel('Order Priority')
plt.ylabel('Count')
plt.legend(title='Segment')
plt.show()
```

The bar plot reveals that the 'Consumer' segment has a higher count of 'High' and 'Medium' order priorities, while 'Corporate' and 'Home Office' segments also frequently have 'Critical' and 'Medium' priorities. This indicates that different customer segments have distinct order priority patterns, which could be relevant for optimizing logistics and customer service strategies.
